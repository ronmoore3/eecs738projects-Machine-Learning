```python
# create a simple gaussian model
from math import exp, pi, sqrt
def gaussian(x, mu, sigma):
    p = []
    N = len(x)
    
    # calculate probability densities for each x
    for i in range(N):
        p.append((1/(sigma * sqrt(2*pi))) * exp((-1/2) *((x[i]-mu)/sigma)**2))
    return p
```

```python
# plot various guassian models with different samples
g1 = gaussian(df['rectal_temp'],38,.5)
g2 = gaussian(df['rectal_temp'],39,.2)
g3 = gaussian(df['rectal_temp'],37,.3)
plt.scatter(df['rectal_temp'], g1, color='b')
plt.scatter(df['rectal_temp'], g2, color='r')
plt.scatter(df['rectal_temp'], g3, color='g')
plt.legend(['g1 = N(38, .5)','g2 = N(39, .2)','g3 = N(37, .3)'], bbox_to_anchor=(1.05, 1), loc=2)
plt.title("Distributions of Various Gaussians")
```

```python
# create a gaussian mixture model using the expectation-maximization (EM) algorithm
from random import choice
def gaussian_mixture_model(x, k):
    # initialize values of means, variances, and mixture coefficients of gaussian models
    N = len(x)
    mix = 1/k
    mu = []
    mu_prev = []
    sig = []
    sig_prev = []
    pi_k = []
    pi_k_prev = []
    lat = []
    gaus = []
    
    # pick good starting values for mu and sig
    used = []
    for i in range(k):
        item = choice(x)
        while item in used:
            item = choice(x)
        used.append(item)
        mu.append(item) 
        sig.append(1.0)
        pi_k.append(mix)
        lat.append([0] * N)
    
    print("intial mu: ", end=" ")
    print(mu)
    print("intial sig: ", end=" ")
    print(sig)
    print("intial pi_k: ", end=" ")
    print(pi_k)
    print(" ")
    
    # initialize guassian variables
    for i in range(k):
        gaus.append(gaussian(x, mu[i],sig[i]))
        
    # repeat E-step and M-step until convergence criterion met
    count = 0
    mu_diff = 1
    sig_diff = 1
    while(mu_diff > .1 and sig_diff > .1):
        # E-step 
        # compute the values of the latent variables
        for i in range(k):
            for j in range(N):
                lat_total = 0
                for l in range(k):
                    lat_total = lat_total + pi_k[l]*gaus[l][j]
                lat[i][j] = pi_k[i]*gaus[i][j]/lat_total
                
        # store prev values of means, variances for convergence testing
        mu_prev = mu
        sig_prev = sig
        pi_k_prev = pi_k
        
        # M-step    
        # update the values of the means, variances, and mixing cofficient
        for i in range(k):
            for j in range(N):
                mu[i] = mu[i] + (x[j] * lat[i][j])/sum(lat[i])
                pi_k[i] = pi_k[i] + lat[i][j]/N
                
        for i in range(k):
            for j in range(N):
                sig[i] = sig[i] + ((x[j]-mu[i])**2 * lat[i][j])/sum(lat[i])
            sig[i] = sqrt(sig[i])
            
        # update gaussian models
        for i in range(k):
            gaus[i] = gaussian(x, mu[i],sig[i])
        
        # increase iteration
        count = count + 1
        
        mu_diff = sum(mu)/len(mu) - sum(mu_prev)/len(mu_prev)
        sig_diff = sum(sig)/len(sig) - sum(sig_prev)/len(sig_prev)
        
#         print("iteration %d completed" % count)
#         print("mu %d:" % count, end=" ")
#         print(mu)
#         print("sig %d:" % count, end=" ")
#         print(sig)
#         print("pi_k %d:" % count, end=" ")
#         print(pi_k)
#         print(' ')
```
